```python
# we use the below function to find more information about the #missing values
def info_missing_table(df_pd):
    """Input pandas dataframe and Return columns with missing value and percentage"""
    mis_val = df_pd.isnull().sum() #count total of null in each columns in dataframe
#count percentage of null in each columns
    mis_val_percent = 100 * df_pd.isnull().sum() / len(df_pd) 
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1) 
 #join to left (as column) between mis_val and mis_val_percent
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'}) 
#rename columns in table
    mis_val_table_ren_columns = mis_val_table_ren_columns[
    mis_val_table_ren_columns.iloc[:,1] != 0].sort_values('% of Total Values', ascending=False).round(1) 
        
    print ("Your selected dataframe has " + str(df_pd.shape[1]) + " columns.\n"    #.shape[1] : just view total columns in dataframe  
    "There are " + str(mis_val_table_ren_columns.shape[0]) +              
    " columns that have missing values.") #.shape[0] : just view total rows in dataframe
    return mis_val_table_ren_columns 
```

```python
kmh_to_mph=0.6213711922
kw_to_hp=1.341022 
mph_to_kmh=1.609344
col_to_clean = "vehicle_desc"

def remove_punctuation(s):
    s = ''.join([i for i in s if i not in set(string.punctuation)])
    return s

def clean_data(df,col_to_clean):
        ## 1) Lower
    df[col_to_clean] = df[col_to_clean].str.lower()
        ## 2) Remove tags
    %time df[col_to_clean] = df.apply(lambda x: re.sub("<[^>]*>", "", x[col_to_clean]), axis=1)
        ## 3) Tokenize
    %time df[col_to_clean+"_proc"] = df.apply(lambda x: word_tokenize(x[col_to_clean]), axis=1)
        ## 4) Remove punctuation
   # table = str.maketrans('', '', string.punctuation)
  #  %time df[col_to_clean+"_proc"] = df.apply(lambda x: [w.translate(table) for w in x[col_to_clean+"_proc"]], axis=1)
    #%time df[col_to_clean+"_proc"] = df[col_to_clean].apply(remove_punctuation)
        ## 5) Remove non-alpha
 #   %time df[col_to_clean+"_proc"] = df.apply(lambda x: [w for w in x[col_to_clean+"_proc"] if w.isalpha()], axis=1)
        ## 6) Remove stop-words
    stop_words = set(stopwords.words('english'))
    %time df[col_to_clean+"_proc"] = df.apply(lambda x: [w for w in x[col_to_clean+"_proc"] if not w in stop_words], axis=1)
        ## 7) Reformat to have a single text. 
    %time df[col_to_clean +"_proc_res"] = df.apply(lambda x: ' '.join(x[col_to_clean+"_proc"]), axis=1)
    
    return df



#we should test Power/Weight Ratio calculation is correct or not!
def preprocess(df,is_kba,col_to_clean,kmh_to_mph,kw_to_hp,mph_to_kmh):
    if is_kba:
        df["id"] = df["Manufacturer Identification Number"].astype(str) + "," + df["Model Identification Numer"].astype(str) 
        df["max. Top Speed/mph"] = df["max. Top Speed"] * kmh_to_mph
        df["max. Top Speed/mph"] =  np.round(df["max. Top Speed"])
        df["max. Top Speed/mph"] = df["max. Top Speed/mph"].astype(int)
        df["Nominal Power/hp"] = df["Nominal Power"] * kw_to_hp
        df["Nominal Power/hp"] =  np.round(df["Nominal Power/hp"])
        df["Nominal Power/hp"] = df["Nominal Power/hp"].astype(int)
        df["fuel_type_desc"] = df["Fuel Code"].apply(lambda x: "Petrol" if x==1 else ("Diesel" if x==2 
                                                                  else ("Electric" if x==4 else ("Petrol LPG" if x==6 
                                                                                               else ("Petrol CNG" if x==7 
                                                                                                    else ("Petrol Electric" if x==8 
                                                                                                         else ("Petrol CNG" if x==9
                                                                                                              else "Diesel Electric" if x==10
                                                                                                              else ("HYDROGEN ELECTRIC" if x==15
                                                                                                                   else ("CNG ELECTRIC" if x==22
                                                                                                                        else ("PETROL ETHANOL" if x==23
                                                                                                                             else ("PETROL ELECTRIC" if x==25
                                                                                                                                  else ("Diesel Electric" if x==26
                                                                                                                                       else ("ETHANOL" if x==34
                                                                                                                                            else ("LNG DIESEL" if x==37
                                                                                                                                                 else "Petrol"))))))))))))))


        df["vehicle_desc"] = df["Manufacturer"].astype(str) + " " + df["Commercial Name"].astype(str) + " " + df["Nominal Power/hp"].astype(str) + "bhp " + df["fuel_type_desc"].astype(str) + ' '  + df["Type"].astype(str)
        df["fuel_type_desc"] = df["fuel_type_desc"].str.lower()
        
    else :
        df["Power/Weight Ratio"] = (df["Power Output"]*mph_to_kmh)/df["Gross Weight"]
    
#         df["fuel_type_desc"] = df["Primary Fuel Type"] + " " + df["Secondary Fuel Type"] --- Bu islem sonucu fueltype bos geliyor
        df["fuel_type_desc"] = df[['Primary Fuel Type', 'Secondary Fuel Type']].apply(lambda x: ' '.join(x.dropna()), axis=1)
        df["fuel_type_desc"] = df["fuel_type_desc"].str.lower()
        df["vehicle_desc"] = df["Manufacturer"] + " " + df["Enhanced Vehicle Description"]
        
    df = clean_data(df,col_to_clean)
    return df
        

%time df_kba_2018_transformed = preprocess(df_kba_2018_2,True,col_to_clean,kmh_to_mph,kw_to_hp,mph_to_kmh)
%time df_abi_transformed = preprocess(df_abi2,False,col_to_clean,kmh_to_mph,kw_to_hp,mph_to_kmh)
```

```python
def fuzzy_match(df_left, df_right, column_left, column_right, threshold=90, limit=1):
    # Create a series
    series_matches = df_left[column_left].apply(
        lambda x: process.extract(x, df_right[column_right], limit=limit)            # Creates a series with id from df_left and column name _column_left_, with _limit_ matches per item
    )

    # Convert matches to a tidy dataframe
    df_matches = series_matches.to_frame()
    df_matches = df_matches.explode(column_left)     # Convert list of matches to rows
    df_matches[
        ['match_string', 'match_score', 'df_right_id']
    ] = pd.DataFrame(df_matches[column_left].tolist(), index=df_matches.index)       # Convert match tuple to columns
    df_matches.drop(column_left, axis=1, inplace=True)      # Drop column of match tuples

    # Reset index, as in creating a tidy dataframe we've introduced multiple rows per id, so that no longer functions well as the index
    if df_matches.index.name:
        index_name = df_matches.index.name     # Stash index name
    else:
        index_name = 'index'        # Default used by pandas
    df_matches.reset_index(inplace=True)
    df_matches.rename(columns={index_name: 'df_left_id'}, inplace=True)       # The previous index has now become a column: rename for ease of reference

    # Drop matches below threshold
    df_matches.drop(
        df_matches.loc[df_matches['match_score'] < threshold].index,
        inplace=True
    )

    return df_matches
```

```python
# function for calculate levenshtein distance
def calculate_distance_levenshtein(df, col1, col2):
    a = df[col1].tolist()
    b = df[col2].tolist()
    score = []
    for i in range(df.shape[0]):
        result = textdistance.levenshtein.normalized_similarity(a[i], b[i])
        score.append(result)
    return score

df_kba_2018_transformed["vehicle_desc_proc_res_dist"] = calculate_distance_levenshtein(df_kba_2018_transformed, col1 = 'vehicle_desc_proc_res', col2 = 'vehicle_desc_proc_res') 
```
